***Métricas***
* ***Visitantes***> Usuários que acessaram qualquer página da plataforma (lista de anúncios, detalhes de um anúncio, conversa de chat, entre outras)


* ***Anunciantes*** >Usuários que tiveram pelo menos um anúncio aprovado


* ***Anúncios aprovados*** >Anúncios que foram inseridos e, por estarem de acordo com as regras da plataforma de classificados online, se tornaram visíveis na lista de anúncios


* ***Receita*** > Receita arrecadada exclusivamente pelos produtos digitais


***Dimenções***
* ***Mês***>Mês em que as métricas foram observadas


* ***Plataforma***>Plataforma digital (Web, Msite ou App) em que o anúncio foi inserido (Anunciantes, Anúncios aprovados, Receita) ou em que a visita à plataforma foi feita (Visitantes)


* ***Vertical***>"Categoria cadastrada no anúncio inserido (Anunciantes, Anúncios aprovados, Receita) ou no anúncio visitado (Visitantes). Se um usuário visita uma lista de anúncios filtrada por determinada Vertical, ele também é contado como visitante dessa vertical.

>**Verticais:**
>* ***Autos*** - automóveis e peças de automóveis


>* ***Imóveis*** - aluguel e venda de imóveis


>* ***Marketplace*** - todo o restante, como eletrônicos, vestimenta, decoração, eletrodomésticos"


***Região***-Região geográfica cadastrada no anúncio inserido (Anunciantes, Anúncios aprovados, Receita) ou no anúncio visitado (Visitantes). Se um usuário visita uma lista de anúncios filtrada por determinada Região, ele também é contado como visitante dessa Região.


***Usuário***-Tipo de usuário (privado ou profissional), como declarado pelo próprio usuário no momento em que criou sua conta na plataforma. Usuários não logados contam como privados, porém, para contatar um vendedor, anunciar ou comprar um produto digital, é necessário ter conta.
